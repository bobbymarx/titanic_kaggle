```python
#convert sex to 0=male, 1=female
x_tmp["Sex"]= np.where(x_tmp['Sex'] == 'male', 0, 1)
```

```python
#set Nan's in Age to mean age
x_tmp['Age'].fillna(x_tmp['Age'].median(), inplace=True)
#normalize fare to min max
scaler = MinMaxScaler()
x_tmp['Fare'] = scaler.fit_transform(x_tmp[['Fare']])
#convert Cabin based on starting letter from 1 to 26 with Nan 27
```

```python
x_tmp
```

```python
#convert Cabin based on starting letter from 1 to 26 with Nan 27
def cabin_to_number(cabin):
    if pd.isna(cabin) or cabin == 'Nan':  # Handling NaN or 'Nan' strings
        return 27
    first_letter = cabin[0]  # Get the first character
    return ord(first_letter.upper()) - ord('A') + 1 if first_letter.isalpha() else 27
```

```python
x_tmp['Cabin'] = x_tmp['Cabin'].apply(cabin_to_number)

print(x_tmp)
```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Sample Training and Test DataFrames
train_df = pd.DataFrame({'Name': ["Eddison, Thomas", "Curie, Marie", "Newton, Isaac"]})
test_df = pd.DataFrame({'Name': ["Einstein, Albert", "Curie, Pierre", "Eddison, Lily"]})

# Step 1: Extract last names
train_df['Last_Name'] = train_df['Name'].str.split(',').str[0]
test_df['Last_Name'] = test_df['Name'].str.split(',').str[0]

# Step 2: Fit LabelEncoder on training data
le = LabelEncoder()
train_df['Last_Name_Encoded'] = le.fit_transform(train_df['Last_Name'])

# Step 3: Create a mapping dictionary
name_mapping = dict(zip(le.classes_, le.transform(le.classes_)))

# Step 4: Transform the test set safely, assigning -1 to unknown values
test_df['Last_Name_Encoded'] = test_df['Last_Name'].map(name_mapping).fillna(-1).astype(int)

print(train_df)
print(test_df)
```

```python
train_df.drop(["Name", "Last_Name"], axis=1, inplace=True)
```
